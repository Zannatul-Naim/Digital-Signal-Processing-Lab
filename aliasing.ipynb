Show that 50Hz is an alias of the frequency 10Hz, when sampling at 40Hz. 

sin(2 pi 10 t) \
sin(2 pi 50 t)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0, 1, 0.001)
x = np.sin(2*np.pi*10*t)

plt.subplot(2, 2, 1)
plt.plot(t, x)
plt.title('x(t) = sin(2*np.pi*10*t)')
plt.xlabel("time(s)")
plt.ylabel('amplitude')

x = np.sin(2*np.pi*50*t)
plt.subplot(2, 2, 2)
plt.plot(t, x)
plt.title('x(t) = sin(2*np.pi*50*t)')
plt.xlabel('time(s)')
plt.ylabel('amplitude')

# sampling frequency
fs = 40
n = np.arange(0, 1, 1/fs)

x_n = np.sin(2*np.pi*10*n)
plt.subplot(2, 2, 3)
plt.stem(n, x_n)
plt.title('x(n) = sin(2*np.pi*10/40*n)')
plt.xlabel('n')
plt.ylabel('amplitude')


x_n = np.sin(2*np.pi*50*n)
plt.subplot(2, 2, 4)
plt.stem(n, x_n)
plt.title('x(n) = sin(2*np.pi*50/40*n)')
plt.xlabel('n')
plt.ylabel('amplitude')

plt.tight_layout()
plt.show()
```
