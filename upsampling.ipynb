```python
import numpy as np
import matplotlib.pyplot as plt

# x = np.sin(2 * np.pi * 0.12 * n)
t = np.arange(0, 0.008, 0.00001)
x = 5 * np.sin(2 * np.pi * 500 * t + np.radians(90))

fs = 5000
n = np.arange(0, 0.008, 1/fs)
x_n = 5* np.sin(2*np.pi*500*n + np.radians(90))

y = np.zeros(3 * len(x_n))
y[0::3] = x_n

plt.figure(figsize=(10, 6))

plt.subplot(4, 1, 1)
plt.plot(t, x)
plt.title('Input Sequence')
plt.xlabel('Time index t(s)')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.stem(n, x_n)
plt.title('Input Sequence')
plt.xlabel('Time index n')
plt.ylabel('Amplitude')


plt.subplot(4, 1, 3)
plt.stem(n, y[:len(x_n)])  # Plot only the length of x
plt.title('Output Sequence (Upsampled)')
plt.xlabel('n -->')
plt.ylabel('Amplitude')

# print(len(n)) = 40
# print(len(x_n)) = 40

# # # down sampling
y_downsampled = x_n[::2]
plt.subplot(4, 1, 4)
plt.stem(np.arange(0, len(n), 2), y_downsampled)
plt.title('Output Sequence (Down Sampled)')
plt.xlabel('n --> ')
plt.ylabel('amplitude')


# Display the plots
plt.tight_layout()
plt.show()
```
