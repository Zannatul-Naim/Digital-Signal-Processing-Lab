```python
import numpy as np
import matplotlib.pyplot as plt
```

# a. Unit Step Sequence

```python
# Define the parameters
N = 21
x = np.ones(N)
n = np.arange(N)
```

```python
# Create the subplot
plt.subplot(2, 2, 1)
plt.stem(n, x)

# Set the labels and title
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Unit Step Sequence')

# Display the plot
plt.show()
```

# b. Exponential sequence

```python
x2 = 0.8 ** n
```

```python
# Create the subplot
plt.subplot(2, 2, 3)
plt.stem(n, x2)

# Set the labels and title
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Exponential Sequence')

# Display the plot
plt.show()
```

# c. Ramp Sequence

```python
# Ask for the length of the ramp sequence
x = int(input('Enter the length of ramp sequence: '))
```

```python
# Generate the ramp sequence
t = np.arange(x+1)
```

```python
# Create the subplot
plt.subplot(2, 2, 1)
plt.stem(t, t)

# Set the labels and title
plt.xlabel('c')
plt.ylabel('Amplitude')
plt.title('Ramp Sequence')

# Display the plot
plt.show()
```

# d. Exponential Sequence (2)

```python
n = np.arange(-10, 11)  # Assuming a range for 'n'

x2 = 0.8 ** n

plt.subplot(2, 2, 3)
plt.stem(n, x2)
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Exponential Sequence')

plt.show()
```

# e. Sinusoidal Sequence

```python
t = np.arange(0, np.pi, 0.01)
y = np.sin(2 * np.pi * t)

plt.subplot(2, 2, 1)
plt.plot(t, y)
plt.ylabel('Amplitude')
plt.xlabel('t')
plt.title('Sinusoidal Sequence')

plt.show()
```

# f. Cosine Sequence

```python
t = np.arange(0, np.pi, 0.01)
y = np.cos(2 * np.pi * t)

plt.subplot(2, 2, 1)
plt.plot(t, y)
plt.ylabel('Amplitude')
plt.xlabel('t')
plt.title('Cosine Sequence')

plt.show()
```
