```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# t = np.arange(0, 1.00, 0.001)
t = np.linspace(0, 2, 1000)
a = 10
f = 5
```

```python
y = a * np.sin(2 * np.pi * f * t)
```

```python
plt.figure(figsize=(10, 8))
plt.subplot(2, 1, 1)
plt.plot(t, y)
plt.grid(True)
plt.xlabel('Time, msec')
plt.ylabel('Amplitude')
plt.title('Continuous-Time Signal x_{a}(t)')
plt.axis([0, 1, -10.2, 10.2])
```

```python
T = 0.01
n = np.arange(0, 1.001, T)
xs = a * np.sin(2 * np.pi * f * n)
k = np.arange(len(n))
```

```python
# plotting the diecrete-time signal

plt.subplot(2, 1, 2)
plt.stem(k, xs, use_line_collection=True)
plt.grid(True)
plt.xlabel('Time index n')
plt.ylabel('Amplitude')
plt.title('Discrete-time signal x[n]')
plt.axis([0, len(n) - 1, -10.2, 10.2])

plt.tight_layout()
plt.show()
```
